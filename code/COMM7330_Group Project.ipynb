### RQ1: What is the current operating status of foreign supermarkets?

### 1.Carrefour in China market

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('../data/'+'Supermarket sales.csv')
df.head(20)
```

```python
df.head()
```

```python
df.dropna(axis=0,how='all')
```

```python
df.rename(columns={"Unnamed: 0": "Years"},inplace=True)
df.head()
```

```python
df.dtypes
```

```python
import matplotlib.ticker as mtick
import matplotlib.pyplot as plt

x1 = df["Years"]
y1 = df["Carrefour Sales (100 million yuan)"]
y2 = df["Number of stores"]
y3 =df["Carrefour growth rate（%）"]
y4 =df["growth rate（%）"]

bar_width=0.3
#set the size of image
plt.rcParams['figure.figsize'] = (12.0,5.0)
fig = plt.figure()
#set the bar chart
plt.bar(x1, height=y1,
    label='Sales', color='steelblue', alpha=0.8, width=bar_width)
plt.bar(x1+ bar_width, height=y2,
    label='Stores', color='indianred', alpha=0.8, width=bar_width)#let % can display
plt.legend(loc='upper left',bbox_to_anchor=(0.29,1))#show legend
plt.ylabel('100 million Yuan', fontsize=8)#set the y label
for a,b,c in zip(x1,y1,y2):
    plt.text(a, b, '%.2f' % b, ha='center', va= 'bottom',fontsize=9)
    plt.text(a+bar_width,c,'%.2f' % c, ha='center', va= 'bottom',fontsize=9)

#set the line chart
ax2 = plt.twinx()
ax2.set_ylabel("growth rate%")
ax2.plot(x1,y3)
ax2.legend(loc = 0)
ax2.plot(x1,y4,color='indianred')#If you want to have two line charts, you only need one ax2, otherwise two axes will appear
plt.title('Carrefour China Market')
plt.figure(figsize=(100,80))
plt.show()
```

 *Data source: CCFA*

The chart below shows the number of stores and turnover in Carrefour China over a period of 5 years. In the first two years, Carrefour showed a downward trend. A significant rise can be sawed in the number of stores and the sum of business from 2015 to 2016.Both the number of stores and turnover peaked in 2016. This is Carrefour's largest and best year in China. After 2016, the number of stores and turnover fluctuated. The growth rate started to show negative growth. Carrefour has gradually fallen into the tide of closing stores and the profit dilemma.

From 2013 to 2017, there were fluctuations. Then, the sum of business remained almost constant from 2017 to 2018.On the other hand, the number of stores is showing a gentle upward trend.
Unlike the negative growth situation of Carrefour, Metro's turnover has been showing positive growth. But even so, in 2018, Metro has also faced a situation where its sales have not increased despite the expansion of its stores. This is also a signal that Metro will face a difficult situation.
